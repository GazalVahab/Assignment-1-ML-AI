## 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
plt.plot(Diabetes_data["Pregnancies"],Diabetes_data["BloodPressure"],"o",color='pink')
plt.xticks(rotation=60)
plt.title("Pregnancy and Blood Pressure")
plt.xlabel("Pregnancies")
plt.ylabel("BloodPressure")
plt.legend(["BloodPressure"],loc="upper right")
plt.grid()
plt.show()
```

```python
import seaborn as sns
```

```python
Diabetes_data.groupby(['Outcome']).sum().plot( 
    kind='pie', y='Pregnancies', autopct='%1.0f%%', shadow=True,colors=sns.color_palette("Set2"),explode=[0,0.10])
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10,6))
plt.bar(Diabetes_data["Pregnancies"],Diabetes_data["Insulin"], color="red",width=0.5)
plt.bar(Diabetes_data["Pregnancies"],Diabetes_data["BloodPressure"], color="Green")
plt.bar(Diabetes_data["Pregnancies"],Diabetes_data["SkinThickness"], color="orange")
plt.xticks(rotation=60)
plt.title("Pregnancies and readings")
plt.xlabel("Readings")
plt.ylabel("Pregnancies")
plt.legend(["Insulin", "Blood Pressure", "Skin Thickness"], loc="upper right")
plt.grid()
plt.show()
```
